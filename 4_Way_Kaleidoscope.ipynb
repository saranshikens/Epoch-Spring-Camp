AIM - To implement a 4-Way kaleidoscope tranformation on a grayscale and an RGB picture

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

**STRATEGY**  
To achieve this transformation, we can see our pic as a matrix,
$\begin{pmatrix}
    P & Q \\
    R & S \\
\end{pmatrix}$.  
Now we choose an arbitrary quadrant say, 'Quadrant Q'.  
We reflect this quadrant about x and y axes as shown in the pic below.


```python
transformation = mpimg.imread('/content/sample_data/Kaleidoscope Transformation.jpg')
plt.axis('off')
plt.imshow(transformation, cmap='gray')
```

A function to convert my RGB picture to grayscale

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

Reading my sample RGB photo

```python
image = mpimg.imread('/content/sample_data/scenery.webp')
```

Storing my converted RGB pic in a separate variable

```python
grayscaleImg = rgb2gray(image)
```

Storing the height and width of the image....it will help in extracting quadrants.

```python
height, width, _ = image.shape
```

**TRANFORMING THE GRAYSCALE IMAGE**

Instead of "Quadrant Q", I have chosen "Quadrant S". This code extracts the bottom right portion of my image.





```python
quadrant_IV = grayscaleImg[height//2:, width//2:]
```

Doing the tranformation

```python
quadrant_II_reflected = np.flipud(quadrant_IV) # Reflection about x-axis
quadrant_III_reflected = np.fliplr(quadrant_IV) # Reflection about y-axis
quadrant_I_reflected = np.flipud(quadrant_III_reflected)

#Combine the quadrants back into a single image
top_half = np.hstack((quadrant_I_reflected, quadrant_II_reflected))  # Combine top-left and top-right
bottom_half = np.hstack((quadrant_III_reflected, quadrant_IV))  # Combine bottom-left and bottom-right

# Final image
kaleidoscope_image = np.vstack((top_half, bottom_half))
```

```python
plt.axis('off')
plt.imshow(kaleidoscope_image, cmap='gray')
```

**TRANFORMING THE RGB IMAGE**

```python
quadrant_IVRGB = image[height//2:, width//2:]
quadrant_II_reflectedRGB = np.flipud(quadrant_IVRGB)
quadrant_III_reflectedRGB = np.fliplr(quadrant_IVRGB)
quadrant_I_reflectedRGB = np.flipud(quadrant_III_reflectedRGB)

top_halfRGB = np.hstack((quadrant_I_reflectedRGB, quadrant_II_reflectedRGB))
bottom_halfRGB = np.hstack((quadrant_III_reflectedRGB, quadrant_IVRGB))

kaleidoscope_imageRGB = np.vstack((top_halfRGB, bottom_halfRGB))
```

```python
plt.axis('off')
plt.imshow(kaleidoscope_imageRGB, cmap='gray')
```
